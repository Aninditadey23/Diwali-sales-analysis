```python
mydata.columns
```

```python
ax = sns.countplot(x="Gender", data= mydata)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_gender=mydata.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
```

```python
df_gender
```

```python
mydata["Gender"].value_counts()
```

```python
sns.barplot(y="Amount",x="Gender", data= df_gender)
```

```python

#from the above bar graph we can interpret that  female spends  more on  products as compared to  males.
```

```python
mydata.columns
```

```python
ax = sns.countplot(x='Age Group',hue="Gender", data= mydata)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=mydata.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders' ,ascending=False)
```

```python
age_gender_spent=mydata.groupby(["Age Group","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Age Group",y="Amount",hue="Gender",data= age_gender_spent)
```

THE AGE GROUP OF 26-35 IS SPENDS MORE THAN THE REST

ORDER OF STATES

```python
sns.barplot(y="State",x="Orders", data= sales_state)
#uttar Pradesh is the city with the highest order and telangana is the lowest.
```

```python
ax = sns.countplot(data = mydata, x = 'Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
FROM THE ABOVE PLOT WE CAN SEE THAT MARRIED PERSONS ARE PURCHASING MORE THAN UNMARRIED
```

```python
sales_states = mydata.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_states
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_states, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
MARRIED FEMALES SPEND MORE THAN UNMARRIED MALES
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.countplot(data = mydata, y = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#majority of buyers are working in the IT Sector and  in the agriculture sector lesser no of people are working
```

```python
sales__state = mydata.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales__state, x = 'Occupation',y= 'Amount')
```

```python
sales_occupation= mydata.groupby(["Occupation","Gender"],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending=False)
```

```python
sales_occupation
```

```python
sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(y="Amount",x="Occupation",hue="Gender",data=sales_occupation)
```

```python
sns.set(rc={'figure.figsize':(20,20)})
ax = sns.countplot(data = mydata, y = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```
